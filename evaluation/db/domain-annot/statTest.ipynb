```python
import numpy as np
import scipy.stats as st
```

```python
X = np.genfromtxt("leftRight-stats.csv",delimiter=";")
```

```python
left = X[:,0]
right = X[:,1]
```

```python
# checking variances ... 
print np.std(left)
print np.std(right)
```

```python
[t, prob] = st.ttest_ind(left, right)
```

```python
if prob < 0.01:
    print "SIGNIFICANT t=", t, " prob = ", prob
else:
    print "insignificant t=", t, " prob = ", prob
```
